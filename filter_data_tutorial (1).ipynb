# Filtrar DataFrames en Pandas

En este tutorial, vamos a ver diferentes estrategias para filtrar datos en un dataframe de Pandas.

En concreto, nos centraremos en los siguientes casos:
- Seleccionar columnas
- Selecccionar filas
- Seleccionar filas y columnas
- Los métodos .loc() y .iloc()
- Filtrar filas en base a condiciones (subset)

## El dataset

```python
# Importamos pandas
import pandas as pd
```

```python
# Vamos a usar la base de datos del titanic. La importamos desde el repositorio abierto de seaborn
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df
```

## Seleccionar columnas

```python
# Seleccionar columna "Name"
df['Name']
```

```python
# Similar a la anterior
df.Name
```

```python
# Si quisiéramos que nos devuelve una columna
df[['Name']]
```

```python
df.columns
```

```python
# Seleccionar varias columnas: Name y Survived
df[['Name','Survived','Pclass']]
```

```python
# Filtrar columnas por tipo de dato
df.info()
```

```python
df.select_dtypes(include = 'int64')
```

```python
df.select_dtypes(include = ['object','int'])
```

```python
# Filtrar columnas que contengan "P"
df.filter(like='P', axis=1)
```

```python
df.filter(regex='^S', axis=1)
```

```python
# Filtrar columnas que contengan "P"
df.filter(regex='ed$', axis=1)
```

## Seleccionar filas

Para seleccionar filas, usamos el índice y dos puntos para indicar el primer el rango de filas que queremos seleccionar. Si no señalamos el principio o el final, Pandas entiende que queremos todas las filas que vienen antes o después.

```python
df[0:1]
```

```python
df[5:10]
```

```python
df[50:]
```

```python
df[:30]
```

```python
df.head(20)
```

```python
df.tail(10)
```

## Seleccionar filas y columnas

```python
# Seleccionar filas y columnas
df[5:10][['Name','Survived']]
```

```python
# Seleccionar columna y filas
df[['Name','Survived']][5:10]
```

```python
df[['Name','Fare']][:]
```

## Los métodos .loc() y iloc()
Eestos métodos nos permiten seleccionar elementos de nuestro dataframe de forma más intuitiva

- .loc() devuelve filas (y/o columnas) con particular labels del índice.

- .iloc() devuelve filas (y/o columnas) en determinadas posiciones en el índice

```python
df.head()
```

```python
# Nombre del pasajero en el índice 2 con .loc()
df.loc[2,'Name']
```

```python
# Nombre del pasajero en el índice 2 con .iloc()
df.iloc[2,3]
```

```python
# Nombre y precio de ticket de pasajeros en índice 100 y 200 con .loc()
df.loc[[100,200],['Name','Fare']]
```

```python
# Nombre y precio de ticket de pasajeros en índice 100 y 200 con .iloc()
df.iloc[[100,200],[3,9]]
```

```python
# Nombre de pasajeros y columnas sucesivas con .loc()
df.loc[:, 'Name':]
```

```python
# Nombre de pasajeros y columnas sucesivas con .iloc()
df.iloc[:, 3:]
```


## 7. Subseting dataframe
También podemos filtrar filas en base a una condición, usando expresiones booleanas


```python
# Condición: Pasajeros en primera clase
df['Pclass'] == 1
```

```python
# Filtrado por pasajeros en primera clase
df[df['Pclass'] ==1]
```

```python
# Pasajeros menores de 18 años
df[df['Age'] < 18].sort_values('Age', ascending=False)
```

```python
# ¿Quién es el pasajero más viejo? 
df[(df['Age'] == df['Age'].max())]
```

```python
# ¿Y el más joven?
df[(df['Age'] == df['Age'].min())]
```

```python
# Filtrado por varias condiciones
df[(df['Sex'] =='female') & (df['Survived'] == 1) & (df['Age'] > 50)]
```

```python
# Filtrado por varias condiciones
df[(df['Name'].str.contains('Miss.')) & (df['Survived'] == 1)]
```
